```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'R:\Cricket Analytics\ipl_data\IPL.csv')
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['year'] = df['date'].dt.year
recent_seasons = [2023, 2024, 2025]
dhoni_recent = df[(df['batter'] == 'MS Dhoni') & (df['year'].isin(recent_seasons))]
match_stats = dhoni_recent.groupby('match_id').agg({
    'runs_batter': 'sum',
    'batter': 'count',
    'striker_out': 'max'
}).rename(columns={'batter': 'balls_faced', 'runs_batter': 'runs_scored'})

total_runs = match_stats['runs_scored'].sum()
total_balls = match_stats['balls_faced'].sum()
innings_out = match_stats['striker_out'].sum()
batting_average = total_runs / innings_out if innings_out != 0 else float('inf')
strike_rate = (total_runs / total_balls) * 100 if total_balls != 0 else 0
innings_played = len(match_stats)
labels = ['Total Runs', 'Balls Faced', 'Strike Rate', 'Average', 'Innings Played', 'Times Out']
values = [total_runs, total_balls, round(strike_rate, 2), round(batting_average, 2), innings_played, int(innings_out)]
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, values, color='skyblue', edgecolor='black')
plt.title("MS Dhoni's IPL Performance (20212023)", fontsize=14)
plt.ylabel("Value")
plt.grid(axis='y', linestyle='--', alpha=0.7)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, round(yval, 2), ha='center', va='bottom')

plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'R:\Cricket Analytics\ipl_data\IPL.csv')
df['date'] = pd.to_datetime(df['date'], errors='coerce')
dhoni_df = df[df['batter'] == 'MS Dhoni'].copy()

# ----- Death Overs -----

death_overs = dhoni_df[(dhoni_df['over'] >= 17) & (dhoni_df['over'] <= 20)]
death_runs = death_overs['runs_batter'].sum()
death_balls = len(death_overs)
death_sr = (death_runs / death_balls) * 100 if death_balls > 0 else 0

# ----- Chasing vs Setting -----

targets = df[df['innings'] == 1].groupby('match_id')['runs_total'].max().reset_index(name='target')
dhoni_df = dhoni_df.merge(targets, on='match_id', how='left')
dhoni_df['batting_first'] = dhoni_df['runs_total'] > dhoni_df['target']

chasing_df = dhoni_df[~dhoni_df['batting_first']]
setting_df = dhoni_df[dhoni_df['batting_first']]

chasing_runs = chasing_df['runs_batter'].sum()
chasing_balls = len(chasing_df)
chasing_sr = (chasing_runs / chasing_balls) * 100 if chasing_balls > 0 else 0

setting_runs = setting_df['runs_batter'].sum()
setting_balls = len(setting_df)
setting_sr = (setting_runs / setting_balls) * 100 if setting_balls > 0 else 0

print("\n5. Clutch Performance Summary")
print(f" Death Overs (1720): {death_runs} runs in {death_balls} balls | SR = {round(death_sr, 2)}")
print(f" Chasing: {chasing_runs} runs in {chasing_balls} balls | SR = {round(chasing_sr, 2)}")
print(f" Setting: {setting_runs} runs in {setting_balls} balls | SR = {round(setting_sr, 2)}")

labels = ['Death Overs (1720)', 'Chasing', 'Setting']
strike_rates = [death_sr, chasing_sr, setting_sr]
colors_bar = ['#ff9933', '#66ccff', '#ff6666']

plt.figure(figsize=(8, 5))
plt.bar(labels, strike_rates, color=colors_bar)
plt.title("MS Dhoni's Clutch Performance (Strike Rate)")
plt.ylabel("Strike Rate")
plt.ylim(0, max(strike_rates) + 20)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

run_labels = ['Death Overs (1720)', 'Chasing', 'Setting']
run_values = [death_runs, chasing_runs, setting_runs]
colors_pie = ['#ffcc99', '#99ccff', '#ff9999']

plt.figure(figsize=(6, 6))
plt.pie(run_values, labels=run_labels, colors=colors_pie, autopct='%1.1f%%', startangle=140)
plt.title("MS Dhoni's Run Contribution by Situation")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load IPL dataset
df = pd.read_csv(r'R:\Cricket Analytics\ipl_data\IPL.csv')
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['season'] = df['date'].dt.year

# Filter only last 3 seasons
recent_df = df[df['season'].isin([2023, 2024, 2025])]

# Filter for CSK batters
csk_recent = recent_df[recent_df['batting_team'] == 'Chennai Super Kings']
csk_batters = csk_recent['batter'].unique()

# Calculate targets for each match
targets = recent_df[recent_df['innings'] == 1].groupby('match_id')['runs_total'].max().reset_index(name='target')

# Function to compute clutch stats
def get_clutch_stats(df, batter):
    player_df = df[df['batter'] == batter].copy()
    
    # Merge target info
    player_df = player_df.merge(targets, on='match_id', how='left')
    player_df['batting_first'] = player_df['runs_total'] > player_df['target']

    # Death overs (1720)
    death = player_df[(player_df['over'] >= 17) & (player_df['over'] <= 20)]
    death_runs = death['runs_batter'].sum()
    death_balls = len(death)
    death_sr = (death_runs / death_balls) * 100 if death_balls > 0 else 0

    # Chasing
    chasing_df = player_df[~player_df['batting_first']]
    chasing_runs = chasing_df['runs_batter'].sum()
    chasing_balls = len(chasing_df)
    chasing_sr = (chasing_runs / chasing_balls) * 100 if chasing_balls > 0 else 0

    # Setting
    setting_df = player_df[player_df['batting_first']]
    setting_runs = setting_df['runs_batter'].sum()
    setting_balls = len(setting_df)
    setting_sr = (setting_runs / setting_balls) * 100 if setting_balls > 0 else 0

    return {
        'batter': batter,
        'death_sr': round(death_sr, 2),
        'chasing_sr': round(chasing_sr, 2),
        'setting_sr': round(setting_sr, 2)
    }

# Compute stats for each CSK batter
results = [get_clutch_stats(recent_df, batter) for batter in csk_batters]
stats_df = pd.DataFrame(results)

# Sort for visual clarity
stats_df = stats_df.sort_values(by='death_sr', ascending=False)

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Death Overs SR
axes[0].barh(stats_df['batter'], stats_df['death_sr'], color='#f9a825')
axes[0].set_title("Death Overs SR (1720)")
axes[0].set_xlabel("Strike Rate")

# Chasing SR
axes[1].barh(stats_df['batter'], stats_df['chasing_sr'], color='#0288d1')
axes[1].set_title("Chasing Strike Rate")
axes[1].set_xlabel("Strike Rate")

# Setting SR
axes[2].barh(stats_df['batter'], stats_df['setting_sr'], color='#43a047')
axes[2].set_title("Setting Strike Rate")
axes[2].set_xlabel("Strike Rate")

plt.suptitle("Clutch Performance: Last 3 Seasons CSK Batters vs MS Dhoni", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv(r'R:\Cricket Analytics\ipl_data\IPL.csv')
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['season'] = df['date'].dt.year

# Filter for last 3 seasons
recent_df = df[df['season'].isin([2023, 2024, 2025])]

# Exclude CSK batters
non_csk_df = recent_df[recent_df['batting_team'] != 'Chennai Super Kings']
non_csk_batters = non_csk_df['batter'].unique()

# Calculate targets
targets = recent_df[recent_df['innings'] == 1].groupby('match_id')['runs_total'].max().reset_index(name='target')

def get_clutch_stats(df, batter):
    player_df = df[df['batter'] == batter].copy()
    player_df = player_df.merge(targets, on='match_id', how='left')
    player_df['batting_first'] = player_df['runs_total'] > player_df['target']

    # Death overs
    death = player_df[(player_df['over'] >= 17) & (player_df['over'] <= 20)]
    death_runs = death['runs_batter'].sum()
    death_balls = len(death)
    death_sr = (death_runs / death_balls) * 100 if death_balls > 0 else 0

    # Chasing
    chasing_df = player_df[~player_df['batting_first']]
    chasing_runs = chasing_df['runs_batter'].sum()
    chasing_balls = len(chasing_df)
    chasing_sr = (chasing_runs / chasing_balls) * 100 if chasing_balls > 0 else 0

    # Setting
    setting_df = player_df[player_df['batting_first']]
    setting_runs = setting_df['runs_batter'].sum()
    setting_balls = len(setting_df)
    setting_sr = (setting_runs / setting_balls) * 100 if setting_balls > 0 else 0

    return {
        'batter': batter,
        'death_sr': round(death_sr, 2),
        'chasing_sr': round(chasing_sr, 2),
        'setting_sr': round(setting_sr, 2)
    }

# Get stats
non_csk_stats = [get_clutch_stats(recent_df, batter) for batter in non_csk_batters]

# Add Dhoni separately
dhoni_stats = get_clutch_stats(recent_df, 'MS Dhoni')
non_csk_stats.append(dhoni_stats)

# Create DataFrame
stats_df = pd.DataFrame(non_csk_stats)
stats_df = stats_df.sort_values(by='death_sr', ascending=False)
stats_df['color'] = stats_df['batter'].apply(lambda x: '#f9a825' if x == 'MS Dhoni' else '#607d8b')

# Plot
fig, axes = plt.subplots(1, 3, figsize=(22, len(stats_df) * 0.35))

# Function for consistent plotting
def plot_bar(ax, values, title):
    ax.barh(stats_df['batter'], values, color=stats_df['color'], edgecolor='black')
    ax.set_title(title, fontsize=14)
    ax.set_xlabel("Strike Rate")
    ax.tick_params(axis='y', labelsize=8)
    ax.grid(axis='x', linestyle='--', alpha=0.4)

# Death Overs
plot_bar(axes[0], stats_df['death_sr'], "Death Overs SR (1720)")

# Chasing
plot_bar(axes[1], stats_df['chasing_sr'], "Chasing SR")

# Setting
plot_bar(axes[2], stats_df['setting_sr'], "Setting SR")

plt.suptitle("MS Dhoni vs Other Batters (Last 3 Seasons)", fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
